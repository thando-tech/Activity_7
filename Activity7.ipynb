```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set_theme()
```

```python
sns.set_style("whitegrid")
sns.set_context("talk")
```

```python
df=pd.read_csv('football.csv')
```

```python
sns.lineplot(x="Goals with right foot", y="Goals with left foot", data=df)
```

```python
sns.distplot(df['Goals with left foot'])
```

```python
sns.kdeplot(data=df['Goals with right foot'], shade=True)
```

```python
sns.rugplot(df['Goals'])
```

```python
sns.boxplot(x='Position', y='Goals', data=df)
```

```python
sns.violinplot(x='Position', y='Headed goals', data=df)
```

```python
sns.swarmplot(x='Position', y='Age', data=df)
```

```python
sns.pairplot(df,hue='Position')
plt.show()
```

```python
sns.jointplot(x='Goals with right foot', y='Goals with left foot', data=df, kind='scatter')
```
